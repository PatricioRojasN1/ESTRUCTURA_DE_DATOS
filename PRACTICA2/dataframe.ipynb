```python
# MEDIA
def media(lista):
    suma = 0
    for valor in lista:
        suma += valor
    return suma / len(lista)
```

```python
def moda(lista):
    frecuencias = {}

    for valor in lista:
        if valor in frecuencias:
            frecuencias[valor] += 1
        else:
            frecuencias[valor] = 1
    
    max_frecuencia = 0
    for frecuencia in frecuencias.values():
        if frecuencia > max_frecuencia:
            max_frecuencia = frecuencia
    
    modas = []
    for valor, frecuencia in frecuencias.items():
        if frecuencia == max_frecuencia:
            modas.append(valor)
    
    return modas
```

```python
# VARIANZA 
def varianza(lista):
    m = media(lista)
    
    suma_cuadrados = 0
    for valor in lista:
        suma_cuadrados += (valor - m) ** 2
    
    return suma_cuadrados / len(lista)
```

```python
# DESVIACIÓN ESTÁNDAR
def desviacion_estandar(lista):
    v = varianza(lista)
    return v ** 0.5

```

```python
columnas_interes = ['price', 'bedrooms', 'bathrooms',
                    'sqft_living', 'sqft_lot', 'floors',
                    'sqft_above', 'sqft_basement', 'yr_built']

for col in columnas_interes:
    print(f"Moda de {col}: {moda(df[col])}")

for col in columnas_interes:
    print(f"Desviación estándar de {col}: {desviacion_estandar(df[col])}")

for col in columnas_interes:
    print(f"Varianza de {col}: {varianza(df[col])}")

for col in columnas_interes:
    print(f"Media de {col}: {media(df[col])}")
```
